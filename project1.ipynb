```python
data = {'Year': [2012, 2013, 2014, 2012, 2013, 2014, 2012, 2013, 2014],
        'Team': ['Birds', 'Birds', 'Birds', 'Deers', 'Deers', 'Deers', 'Caps',
                 'Caps', 'Caps'],
        'Wins': [4, 5, 10, 14, 11, 6, 7, 4, 6],
        'Losses': [5, 6, 7, 1, 5, 12, 4, 11, 8]}
```

```python
import numpy as np
import pandas as pd

df = volleyball
df
```

```python
#Adding column with number of games played


df['Games_Played'] = df.Wins + df.Losses
df
```
